```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def forward(fx,h):
    """
    Calculate the forward difference
    """
    dfdx = np.zeros(len(fx))
    for i,f in enumerate(fx[:-1]):
        dfdx[i] = (fx[i+1]-fx[i])/h
    return dfdx
```

```python
def backward(fx,h):
    """
    Calculate the backward difference
    """
    dfdx = np.zeros(len(fx))
    for i in range(1,len(fx)):
        dfdx[i] = (fx[i]-fx[i-1])/h
    return dfdx
```

```python
def center(fx,h):
    """
    Calculate the centered difference
    """  
    dfdx = np.zeros(len(fx))
    for i in range(1,len(fx)-1):
        dfdx[i] = (fx[i+1]-fx[i-1])/(2.*h)
    return dfdx
```

```python
func  = lambda x: np.sin(x)
dfunc = lambda x: np.cos(x)
h  = 0.1
xx = np.linspace(0,1,(int(1/h)+1))
yy = func(xx) 
dx = xx[2]-xx[1]
```

```python
print(xx,len(xx))
```

```python
dydx = dfunc(xx)
```

```python
dydx_forward  = forward(func(xx),h=h)
dydx_backward = backward(func(xx),h=h)
dydx_center   = center(func(xx),h=h)
```

```python
plt.figure(figsize=(8,6))
plt.plot(xx,dydx,'k-',label="True answer")
plt.plot(xx[:-1],dydx_forward[:-1],label="Forward")
plt.plot(xx[1:],dydx_backward[1:],label="Backward")
plt.plot(xx[1:-1],dydx_center[1:-1],label="Centered")
plt.legend(loc="best")
plt.xlim([0,1])
plt.ylim([0.5,1.1])
```

```python
plt.figure(figsize=(8,6))
plt.plot(xx,dydx,'k-',label="True answer")
plt.plot(xx[:-1]+0.5*dx,dydx_forward[:-1],label="Forward")
plt.plot(xx[1:]-0.5*dx,dydx_backward[1:],label="Backward")
plt.plot(xx[1:-1],dydx_center[1:-1],label="Centered")
plt.legend(loc="best")
plt.xlim([0,1])
plt.ylim([0.5,1.1])
```
